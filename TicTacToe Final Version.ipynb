```python
from tkinter import * 
from tkinter import messagebox
import random
```

```python
def next_turn(row,col):
    global player
    if matrix[row][col]['text']=="" and check_winner()==False:
        if player==players[0]:
            matrix[row][col]['text']=player
            if check_winner()==False:
                player=players[1]
                label.config(text=(players[1]+ " Turn"))
            elif check_winner()==True:
                 label.config(text=(players[0]+ " Wins"))
                 update_scores(players[0])
                 messagebox.showinfo("Congrats!",players[0] + " Won")
            elif check_winner()=='Tie':           
                 label.config(text=("Tie!"))
        elif player==players[1]:
            matrix[row][col]['text']=player
            if check_winner()==False:
                player=players[0]
                label.config(text=(players[0]+ " Turn"))
            elif check_winner()==True:
                 label.config(text=(players[1]+ " Wins"))
                 update_scores(players[1])
                 messagebox.showinfo("Congrats!",players[1] + " Won")
            elif check_winner()=='Tie':           
                 label.config(text=("Tie !"))
           
def update_scores(winner):
    
    global x_wins
    global o_wins
    
    if winner == "X":
        x_wins += 1
    if winner == "O":
        o_wins += 1
        
    scores.config(text=("X Score: " + str(x_wins) + "\nO Score: " + str(o_wins)))
    
def check_winner():
    #Check for all rows
    for row in range(3):
        if matrix[row][0]['text']==matrix[row][1]['text']==matrix[row][2]['text']!="":
            return True
    #Check for all columns
    for col in range(3):
        if matrix[0][col]['text']==matrix[1][col]['text']==matrix[2][col]['text']!="":
            return True
    #Checking the diagonals
    if matrix[0][0]['text']==matrix[1][1]['text']==matrix[2][2]['text']!="":
            return True
    elif matrix[0][2]['text']==matrix[1][1]['text']==matrix[2][0]['text']!="":
            return True
    #Check Tie
    if check_empty_spaces()==False:
        return "Tie"
    else:
        return False

def check_empty_spaces():
    spaces = 9
    for row in range(3):
        for col in range(3):
            if matrix[row][col]['text'] != "":
                spaces -= 1
    if spaces == 0:
        return False
    else:
        return True

def start_new_game():
    global player
    player=random.choice(players)
    label.config(text=(player + " Turn"))
    for row in range(3):
            for col in range(3):
                matrix[row][col].config(text="")
```

```python
#colors
grid_color="#3D3B40"
player_color="#525CEB"
window_color="#BFCFE7"
text_color="#F8EDFF"

window=Tk()
window.title("Tic Tac Toe")
window.geometry("485x530")
window.maxsize(485,530)
window.configure(bg=window_color)

#List of Players,which are X and O
players=['X','O']
#Choose random player either X or O , when starting the game
player=random.choice(players)

#score variables
x_wins = 0
o_wins = 0

#Creating 3x3 matrix 
matrix=[[0,0,0],[0,0,0],[0,0,0]]

label_frame = Frame(window, width=100,bg=window_color)
label_frame.pack(side="top")

# Scores label
scores = Label(label_frame, text="X Score: " + str(x_wins) + "\nO Score: " + str(o_wins), font=('Arial', 15, 'bold'),bg=window_color, fg=text_color)
scores.grid(row=0, column=0, padx=28, sticky="nsew")  # Added sticky parameter for equal spacing

# Restart Button
restart_button = Button(label_frame, text="New Game", font=('Arial', 20, 'bold'), bg=player_color,fg=text_color, command=start_new_game)
restart_button.grid(row=0, column=1, padx=3, sticky="nsew")  # Added sticky parameter for equal spacing

# Player turn label
label = Label(label_frame,bg=window_color,fg=text_color, text=(player + " Turn"), font=('Arial', 30,'bold'))
label.grid(row=0, column=2, padx=15, sticky="nsew")  # Added sticky parameter for equal spacing

# Configure column weights
label_frame.grid_columnconfigure(0, weight=1)
label_frame.grid_columnconfigure(1, weight=1)
label_frame.grid_columnconfigure(2, weight=1)


#Creating matrix buttons
buttons_frame=Frame(window, bg=grid_color)
buttons_frame.pack()

for row in range(3):
    for col in range(3):
        matrix[row][col]=Button(buttons_frame,text="",font=('Calibri', 55, "bold"),bg= grid_color,fg=player_color,width=4,height=1,command=lambda row=row,col=col: next_turn(row,col))
        matrix[row][col].grid(row=row,column=col)
        
window.mainloop()
```
